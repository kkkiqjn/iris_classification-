Classificatio of four iris species using KNN 

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import mglearn
import numpy as np
```

```python
iris_dataset=load_iris()
```

```python
print(iris_dataset["DESCR"])
```

```python
dframe=pd.DataFrame(iris_dataset["data"])
dframe.columns=iris_dataset["feature_names"]
print(dframe.head())
```

```python
X_train,X_test,y_train,y_test=train_test_split(iris_dataset["data"],iris_dataset["target"],random_state=0)
```

```python
print(X_train.shape\
,y_train.shape)
```

```python
print(X_test.shape,y_test.shape)
```

```python
iris_dframe=pd.DataFrame(X_train,columns=iris_dataset["feature_names"])
grr=pd.plotting.scatter_matrix(iris_dframe,c=y_train,figsize=(16, 15), marker='o',
 hist_kwds={'bins':20}, s=50, alpha=.9, cmap=mglearn.cm3)
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
y=knn.predict([X_test[0]])
print(iris_dataset["target_names"][y[0]])
```

```python
y_pred=knn.predict(X_test)
acc=np.mean(y_test==y_pred)
print("accuracy:{:.2f}".format(acc))
```
